```python
# Common Imports & Polite Headers
import re, sys, pandas as pd, requests
from bs4 import BeautifulSoup
HEADERS = {"User-Agent": (
    "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 "
    "(KHTML, like Gecko) Chrome/122.0 Safari/537.36")}
def fetch_html(url: str, timeout: int = 20) -> str:
    r = requests.get(url, headers=HEADERS, timeout=timeout)
    r.raise_for_status()
    return r.text
def flatten_headers(df: pd.DataFrame) -> pd.DataFrame:
    if isinstance(df.columns, pd.MultiIndex):
        df.columns = [" ".join([str(x) for x in tup if str(x)!="nan"]).strip()
                      for tup in df.columns.values]
    else:
        df.columns = [str(c).strip() for c in df.columns]
    return df
print("Common helpers loaded.")
```

```python
def q2_clean(df: pd.DataFrame) -> pd.DataFrame:
    for col in ['rank', 'points', 'comments']:
        df[col] = df[col].astype(str).str.extract(r'(\d+)').fillna('0').astype(int)

    df[['title', 'link', 'user']] = df[['title', 'link', 'user']].fillna('')

    return df

def q2_parse_items(html: str) -> pd.DataFrame:
    soup = BeautifulSoup(html, 'lxml')
    data = []

    story_rows = soup.find_all('tr', class_='athing')

    for row in story_rows:
        story_data = {'rank': '0', 'title': '', 'link': '', 'points': '0', 'comments': '0', 'user': ''}

        rank_tag = row.find('span', class_='rank')
        if rank_tag:
            story_data['rank'] = rank_tag.text.strip('.')

        title_link = row.find('span', class_='titleline').find('a')
        if title_link:
            story_data['title'] = title_link.text
            story_data['link'] = title_link.get('href', '')

        subtext_row = row.find_next_sibling('tr')
        subtext_tag = subtext_row.find('td', class_='subtext') if subtext_row else None

        if subtext_tag:
            score_tag = subtext_tag.find('span', class_='score')
            if score_tag:
                story_data['points'] = score_tag.text.split(' ')[0]

            user_tag = subtext_tag.find('a', class_='hnuser')
            if user_tag:
                story_data['user'] = user_tag.text

            links = subtext_tag.find_all('a')

            comments_text = next((a.text for a in links if 'comment' in a.text or a.text == 'discuss'), '0')

            if comments_text == 'discuss':
                story_data['comments'] = '0'
            else:
                match = re.match(r"(\d+)", comments_text)
                if match:
                    story_data['comments'] = match.group(1)
                else:
                    story_data['comments'] = '0'

        data.append(story_data)

    return pd.DataFrame(data)

def q2_sort_top(df: pd.DataFrame, top: int = 15) -> pd.DataFrame:
    df_sorted = df.sort_values(by='points', ascending=False).head(top)

    return df_sorted
```
